```python
import sqlite3
import pandas as pd

# Create sample data
data = {
    'order_id': [1,2,3,4,5,6,7],
    'order_date': ['2023-01-10','2023-01-15','2023-02-05','2023-02-20','2023-03-01','2023-03-10','2023-03-15'],
    'amount': [100,200,150,300,250,400,350],
    'product_id': [101,102,103,101,102,103,104]
}

df = pd.DataFrame(data)

# Create SQLite DB
conn = sqlite3.connect(':memory:')
df.to_sql('online_sales', conn, index=False, if_exists='replace')

df.head()
```

```python
from google.colab import sheets
sheet = sheets.InteractiveSheet(df=df)
```

```python
query = """
SELECT
    strftime('%Y', order_date) AS year,
    strftime('%m', order_date) AS month,
    SUM(amount) AS total_revenue,
    COUNT(DISTINCT order_id) AS total_orders
FROM online_sales
GROUP BY year, month
ORDER BY year, month;
"""

result = pd.read_sql(query, conn)
result
```

```python
from google.colab import sheets
sheet = sheets.InteractiveSheet(df=result)
```

```python
query2 = """
SELECT
    strftime('%Y', order_date) AS year,
    strftime('%m', order_date) AS month,
    SUM(amount) AS total_revenue
FROM online_sales
GROUP BY year, month
ORDER BY total_revenue DESC
LIMIT 3;
"""

pd.read_sql(query2, conn)
```
